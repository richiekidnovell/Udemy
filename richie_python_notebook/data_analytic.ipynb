```python
edx_course_description['n_enrolled'] = edx_course_description['n_enrolled'].str.replace(',', '')
edx_course_description['price'] = edx_course_description['price'].str.replace('FREE-Add a Verified Certificate for ', '')
edx_course_description['price'] = edx_course_description['price'].str.replace('$', '', regex=False)
edx_course_description['price'] = edx_course_description['price'].str.replace('USD', '')
edx_course_description['course_length'] = edx_course_description['course_length'].str.replace('Weeks', '')
```
